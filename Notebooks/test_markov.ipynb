```python
import numpy as np

dt = 1.
t_end = 100
def t_vec():
    return np.arange(0,t_end,dt)
def y_vec(t_vec, cov = 1.):
    M = np.zeros(len(t_vec))
    for i in range(1,len(M)):
        M[i] = M[i-1]+(2*np.random.rand()-1)*cov
    return M
        

t1 = t_vec()
y1 = y_vec(t1,1)


import matplotlib.pyplot as plt
for i in range(100):
    plt.plot(t1, y_vec(t1,1))
# plt.plot(<X AXIS VALUES HERE>, <Y AXIS VALUES HERE>, 'line type', label='label here')
# plt.plot(<X AXIS VALUES HERE>, <Y AXIS VALUES HERE>, 'line type', label='label here')
# plt.legend(loc='best')
plt.show()
```
