```python
%%cython -a

cdef double lin_fun(double x, double k):
    return x*k

cdef double fun(double (*fun_type)(double, double), double a, double b):
    return fun_type(a, b);

cdef double lin(double *x):
    return x[0]

print(fun(&lin_fun,10,2))

def sens(fun_type, double x):
    if fun_type=='linear':
        return lin(&x, )
    
print(sens('linear',10))
```

```python
%%cython -a
cimport numpy as np
import numpy as np
from cythonic.plugins.positions cimport point
from libc.math cimport lrint # round double, cast as long

cdef class MeshMap:
    cdef point dim
    cdef double pitch
    cdef double[:,::1] map
    def __cinit__(self, double[:] dim, double resolution):
        """ base class for a meshgrid and some supporting functions """
        " assume dim is vector[2]=> [double x, double y]"
        self.dim = point(dim[0], dim[1])
        self.pitch = resolution
        self.map = np.ones((100,100),dtype = np.float_)

    cpdef long to_grid(self, double  x):
        " round point in mm to grid index "
        return lrint(x/self.pitch)
    
M = MeshMap(np.array([100,100],dtype=np.float_), .25)
cdef double x = 100
# cdef long y = M.to_grid(&x)
cdef do_something():
    cdef double XX = 100.
    print(M.to_grid(XX))
    
do_something()
cdef double xx 
cdef long y = 10
cdef long z = 11
xx = <double>y+z
print(xx)
```
